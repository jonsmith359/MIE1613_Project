```python
stocks ={
    'A':{'S0':100., 'mu':0.05, 'sigma':0.1},
    'B':{'S0':50., 'mu':0.1, 'sigma':0.2},
#     'C':{'S0':20., 'mu':0.08, 'sigma':0.12},
#     'D':{'S0':80., 'mu':0.02, 'sigma':0.03},
#     'E':{'S0':120., 'mu':0.05, 'sigma':0.2},
    }
r = 0.05
T = 1
reps = 1000
steps = 100
```

# Long 1 stock

```python
portf = OptionPortfolio(stocks = stocks,r=r, T=T, reps=reps, steps=steps, sensitivity=True, delta=0.1, stock='A')
```

```python
portf.add_call(stock='A',num=5,K=100,sense='buy',op_type='european')
portf.add_call(stock='A',num=5,K=120,sense='sell',op_type='european')
portf.add_stock(stock='A',num=1)
print portf.products
portf.stock_plot('A')
```
