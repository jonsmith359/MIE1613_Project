```python
# Standard Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Custom Libraries
from Option_Portfolio import OptionPortfolio
```

# Stock Parameters

```python
stocks ={
    'A':{'S0':100., 'mu':0.05, 'sigma':0.1},
    'B':{'S0':50., 'mu':0.1, 'sigma':0.2},
    'C':{'S0':20., 'mu':0.03, 'sigma':0.12},
    }
r = 0.05
T = 1
reps = 1000
steps = 100
delta = 0.1
```

```python
portf = OptionPortfolio(stocks = stocks,r=r, T=T, reps=reps, steps=steps, sensitivity=True, delta=delta, stock='A')
portf.add_stock(stock='A',num=5, sense='long')
portf.add_put(stock='A',num=5,K=120,sense='buy',op_type='european')
portf.add_put(stock='A',num=5,K=100,sense='sell',op_type='european')

portf.add_call(stock='B',num=5,K=50,sense='buy',op_type='american', exercise=0.5)
portf.add_call(stock='B',num=5,K=55,sense='sell',op_type='american', exercise=0.5)

portf.add_stock(stock='C',num=5, sense='short')

sens_port_list,sens_port_ave,sens_port_ci,FD_list,FD_ave,FD_ci = portf.sensitivity_analysis()
```

```python
print '------------\nDelta=',0.1
print '95% Confidence Interval of Derivative:',FD_ave,'+/-',FD_ci
```

```python
print 'Average Net portfolio value:', np.mean(portf.net_value()[0])
print 'List of Portfolio Products'.center(85,"-")
print portf.products
print '-'*85
portf.hist_plot()
```
